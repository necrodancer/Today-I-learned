## 주석 달기

(1) 정답 코드

```python
# import modules, tensorflow, numpy
import tensorflow as tf
import numpy as np

# load csv data with numpy
data = np.loadtxt('./data.csv', delimiter=',',
                  unpack=True, dtype='float32')

# transpose x_data and y_data to train
# data spec : index 0,1,2 = attributes
#             index 2,3 = label, one-hot encoded
x_data = np.transpose(data[0:2])
y_data = np.transpose(data[2:])

#########
# Implementation of neural network
######
# define global_step not included in training, named 'global_step'
global_step = tf.Variable(0, trainable=False, name='global_step')

# define placeholder X and Y with float32 type.
X = tf.placeholder(tf.float32)
Y = tf.placeholder(tf.float32)

# define 'layer1' with tf.name_scope
with tf.name_scope('layer1'):
    # input shape is [2,10]. initialize with random_uniform, named W1.
    # use relu.
    W1 = tf.Variable(tf.random_uniform([2, 10], -1., 1.), name='W1')
    L1 = tf.nn.relu(tf.matmul(X, W1))
    # get histogram of 'X'(X) and 'Weights'(W1).
    tf.summary.histogram("X", X)
    tf.summary.histogram("Weights", W1)
# define 'layer2' with tf.name_scope
with tf.name_scope('layer2'):
    # shape is [10,20], initialize with random_uniform, named W2
    W2 = tf.Variable(tf.random_uniform([10, 20], -1., 1.), name='W2')
    L2 = tf.nn.relu(tf.matmul(L1, W2))
    # get histogram of 'Weights'(W2).
    tf.summary.histogram("Weights", W2)
# define 'output' with tf.name_scope
with tf.name_scope('output'):
    # shape is [20, 3]
    # not use activation function, just matmul.
    W3 = tf.Variable(tf.random_uniform([20, 3], -1., 1.), name='W3')
    model = tf.matmul(L2, W3)
    # get histogram aof Weights-W3 and Model-model
    tf.summary.histogram("Weights", W3)
    tf.summary.histogram("Model", model)
# define 'optimizer' with tf.name_scope
with tf.name_scope('optimizer'):
    # define cost using softmax_cross_entropy_with_logits
    cost = tf.reduce_mean(
        tf.nn.softmax_cross_entropy_with_logits(labels=Y, logits=model))
    # define optimizer using AdamOptimizer, lr = 0.01
    optimizer = tf.train.AdamOptimizer(learning_rate=0.01)
    # define train_op and minimize with global_step
    train_op = optimizer.minimize(cost, global_step=global_step)
    
    # get summary scalar of 'cost'(cost).
    tf.summary.scalar('cost', cost)

#########
# Training the neural network
######
# define session and saver
sess = tf.Session()
saver = tf.train.Saver(tf.global_variables())

# define check point using get_checkpint_state()
ckpt = tf.train.get_checkpoint_state('./model')
# if ckpt and is_ckpt_exist == True
#     restore
# else
#     run global_variables_initializer
if ckpt and tf.train.checkpoint_exists(ckpt.model_checkpoint_path):
    saver.restore(sess, ckpt.model_checkpoint_path)
else:
    sess.run(tf.global_variables_initializer())

# define merged and writer
merged = tf.summary.merge_all()
writer = tf.summary.FileWriter('./logs', sess.graph)

# Training with 100 steps
for step in range(100):
    # run train_op
    sess.run(train_op, feed_dict={X: x_data, Y: y_data})
    
    # print step and cost
    print('Step: %d, ' % sess.run(global_step),
          'Cost: %.3f' % sess.run(cost, feed_dict={X: x_data, Y: y_data}))
    
    # define summary running merged
    # Do add_summary
    summary = sess.run(merged, feed_dict={X: x_data, Y: y_data})
    writer.add_summary(summary, global_step=sess.run(global_step))

# save the model trained
saver.save(sess, './model/dnn.ckpt', global_step=global_step)

#########
# Testing the neural network
######
# define prediction with argmax
# define target with argmax
prediction = tf.argmax(model, 1)
target = tf.argmax(Y, 1)
# print predicted_value
# print target_value
print('예측값:', sess.run(prediction, feed_dict={X: x_data}))
print('실제값:', sess.run(target, feed_dict={Y: y_data}))

# define is_correct with tf.equal
is_correct = tf.equal(prediction, target)
# define accuracy using tf.reduce_mean and tf.cast
# print accuracy as .2f
accuracy = tf.reduce_mean(tf.cast(is_correct, tf.float32))
print('정확도: %.2f' % sess.run(accuracy * 100, feed_dict={X: x_data, Y: y_data}))
```
